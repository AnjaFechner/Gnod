# Create lists for songs, artists and genre

```python
# I limit the len, because it is to much for my laptop. It needs 40 minutes for 800 songs.
songs = []
artists = []
genres = []

len_song = 500

for i in tqdm(range(len_song)):
    songs.append(soup.select("td.table1column2 > p > span")[i].text)
    artists.append(soup.select("td.table1column4 > p > span")[i].text)
    genres.append(soup.select("td.table1column3 > p > span")[i].text)
```

# Create the DataFrame 

```python
file1 = pd.DataFrame({'song': songs, 'artists': artists, 'genre': genres})
```
