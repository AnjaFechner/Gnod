- Iterations the scrap the next pages

```python
# I limit the number of pages by 50 of 345 (content of every page: 200 songs)
iterations = range(1, 50)
```

```python
pages = []
url_list = []

for i in iterations:
    start_at = str(i)
    url = f"https://freemusicarchive.org/music/charts/this-month?pageSize=200&page={start_at}"
    url_list.append(url)
    response = requests.get(url)
    print("Status code: " + str(response.status_code))
    pages.append(response)
    wait_time = randint(1,4)
    sleep(wait_time)
url_list    
```

```python
soup = BeautifulSoup(pages[0].content, "html.parser")
```

2. Scraping the content

- song

```python
#Chrome
soup.select("span.chartcol-track > a > b")
```

```python
#Firefox
soup.select("div.play-item > span:nth-child(4)")
```

- artist

```python
#Chrome
soup.select("span.chartcol-artist > a")
```

```python
#Firefox
soup.select("div.play-item > span:nth-child(3)")
```

- genre

```python
#Chrome
soup.select("span.chartcol-genre > a:nth-child(1) > b")
```

```python
# Firefox
soup.select("div.play-item > span:nth-child(6)")
```

3. Create a list for song, artist and genre

```python
len(pages)
```

```python
pages_parsed = []

# looping through all pages
for i in range(len(pages)):
    # a soup for ever search result page
    pages_parsed.append(BeautifulSoup(pages[i].content, "html.parser"))
```

```python
len(pages_parsed)
```

- Songs

```python
songs = []
artists = []
genres = []

for i in range(len(pages)):
    song_html = pages_parsed[i].select("div.play-item")
   
    for j in range(len(song_html)):
        songs.append(song_html[j].select("span:nth-child(4)")[0].get_text())
        
        artists.append(song_html[j].select("span:nth-child(3)")[0].get_text())
        
        genres.append(song_html[j].select("span:nth-child(6)")[0].get_text())
```

```python
len(song_html)
```

- 4. Create the Dataframe

```python
file2 = pd.DataFrame({'song': songs, 'artists': artists, 'genre': genres})
```

```python
file2
```

- 5. Delete of not necesdsary signs in the list

```python
file2['song'] = file2['song'].apply(lambda x: x.lstrip('\n').strip())
```

```python
file2['artists'] = file2['artists'].apply(lambda x: x.lstrip('\n').strip())
```

```python
file2['artists'] = file2['artists'].apply(lambda x: x.lstrip('\n').strip())
```
