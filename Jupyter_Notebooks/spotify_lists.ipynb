0. Libraries

```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import getpass
import pandas as pd
```

1. Connecting to Spotify

```python
client_id = str(getpass.getpass('client_id?'))
client_secret = str(getpass.getpass('client_secret?'))
```

```python
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id=client_id,
                                                           client_secret=client_secret))
```

2. Choose a playlist

```python
Top_of_the_Morning = sp.user_playlist_tracks("Spotify", "37i9dQZF1DWSWyJydK4fTU")
```

- how many songs are in the playlist

```python
len(Top_of_the_Morning['items'])
```

3. to get the id per song

```python
# let's get the ID for only one song first!
Top_of_the_Morning['items'][0]['track']['id']

# here i make it work for the whole playlist
for item in Top_of_the_Morning['items']:
    print(item['track']['id'])

playlist = []
for item in Top_of_the_Morning['items']:
    playlist.append(item['track']['id'])
```

```python
sp.audio_features(tracks=playlist)
```

```python
audio_features_Top_of_The_Morning = pd.DataFrame(sp.audio_features(tracks=playlist))
```

```python
audio_features_Top_of_The_Morning.head()
```

4. to get the name and the artist of the songs

```python
Top_of_the_Morning["items"][0]["track"]["name"]
```

- song

```python
# let's get the ID for only one song first!
Top_of_the_Morning['items'][0]['track']["name"]

# here i make it work for the whole playlist
for item in Top_of_the_Morning['items']:
    print(item['track']['name'])
    
playlist1 = []
for item in Top_of_the_Morning['items']:
    playlist1.append(item['track']['name'])
```

- the artists

```python
Top_of_the_Morning['items'][0]['track']['artists'][0]['name']
```

```python
# let's get the ID for only one song first!
Top_of_the_Morning['items'][0]['track']['artists'][0]['name']

# here i make it work for the whole playlist
for item in Top_of_the_Morning['items']:
    print(item['track']['artists'][0]['name'])

playlist2 = []
for item in Top_of_the_Morning['items']:
    playlist2.append(item['track']['artists'][0]['name'])
```

5. create new DataFrames

```python
song_Top_of_The_Morning = pd.DataFrame(playlist1, columns = ["song"])
artists_Top_of_The_Morning = pd.DataFrame(playlist2, columns = ["artists"])
```

```python
song_Top_of_The_Morning.head()
```

```python
len(song_Top_of_The_Morning), len(artists_Top_of_The_Morning), len(audio_features_Top_of_The_Morning)
```

```python
spotify_list = pd.concat((song_Top_of_The_Morning, artists_Top_of_The_Morning,audio_features_Top_of_The_Morning), axis=1 )
```
